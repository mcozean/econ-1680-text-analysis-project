```python
# Display the top 10 skills 
top_N = 10 

# Split the comma-separated skills and generate a list of all skills
all_skills = [skill.strip("'[] ") for sublist in df['Skills'].dropna().str.split(',') for skill in sublist]

# Count the skills in all_skills array and display top skills in a table with skill and count
top_skills = pd.DataFrame(Counter(all_skills).most_common(top_N), columns=['Skill','Count']).set_index('Skill')
print(top_skills)
```

```python
# Strip skills 
df['Skills'] = df['Skills'].str.strip()

# Create a string containing all skills separated by underscores
all_skills = ' '.join([skill.strip("'[] ").replace(' ', '_') for sublist in df['Skills'].dropna().tolist() for skill in sublist.split(',')])

# Generate word cloud
all_skills_word_cloud = WordCloud(collocations=False, background_color='white').generate(all_skills)

# Display image with bilinear interpolation
plt.imshow(all_skills_word_cloud, interpolation='bilinear')

# Save word cloud for all skills in output path
all_skills_word_cloud.to_file(os.path.join(output_path,"allskills_wordcloud.png"))

# Remove axis
plt.axis("off")

# Show plot
plt.show()
```

```python
# Modify the skills strings to add underscores between spaces only when a skill spans multiple words
df['Skills'] = df['Skills'].apply(lambda x: re.sub(r'\s+', '_', x))

# Initialize term frequency vectorizer called vectorizer
vectorizer = TfidfVectorizer()

# Tokenize each skill string by splitting at commas
X = vectorizer.fit_transform(df['Skills'])

# Print the dimensions of the document-term matrix X
print(X.shape)

# Obtain list of skills
features = vectorizer.get_feature_names_out()

# Print the list of skills (features)
print(features)

# Generate binary output variable for regression using cutoff
y = (df['Average Salary Estimate'] > 104500)

# Generate input variable for regression
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1680)

# Initialize empty dictionary of test scores
test_scores = {}

# Set maximum score at 0
max_score = 0

# Loop through various hyperparameters c
for c in [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000]:
    # Run logistic regression with c
    loglasso = LogisticRegression(penalty='l1', solver='liblinear', C=c, random_state=1680).fit(x_train, y_train)
    
    # Calculate test score
    test_score = loglasso.score(x_test, y_test)
    
    # Check if test score > max score
    if test_score > max_score:
        # Set best model to log lasso
        best_model = loglasso
        # Set max score to test score
        max_score = test_score
        
    # Add c and corresponding test score to dictionary
    test_scores[c] = test_score

# Print dictionary of test scores
print(test_scores)

# Extract coefficients from the best model
coefs = best_model.coef_

# Create dataframe with skills and coefficients
df_coef = pd.DataFrame({'Skill': features, 'Coef': coefs[-1].tolist()})

# Create column in df_coef with absolute value of coefficient
df_coef['Abs Coef'] = df_coef['Coef'].abs()

# Sort dataframe by absolute value of coefficients
print(df_coef.sort_values(by='Abs Coef', ascending=False).head(60))

# Create confusion matrix
cm = confusion_matrix(y_test, best_model.predict(x_test))

# Initialize confusion matrix display object
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=best_model.classes_)

# Plot confusion matrix
disp.plot()

# Save plot in output path
plt.savefig(os.path.join(output_path, 'confusionmatrix_allskills.png'))

# Show plot
plt.show()
```
